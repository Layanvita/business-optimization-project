```python
# Import PuLP library
from pulp import *

# Create optimization problem
problem = LpProblem(
    "Profit_Maximization",
    LpMaximize
)

print("Problem Created Successfully")

# Number of units to produce
A = LpVariable(
    "Product_A",
    lowBound=0,
    cat="Integer"
)

B = LpVariable(
    "Product_B",
    lowBound=0,
    cat="Integer"
)

print("Decision Variables Created")

# Maximize Profit
problem += (
    40 * A +
    30 * B
)

print("Objective Function Added")

# Labor constraint
problem += (
    2 * A +
    1 * B <= 100
)

# Machine constraint

problem += (
    1 * A +
    1 * B <= 80
)

print("Constraints Added")

# Solve optimization problem
problem.solve()

print(
    "Status:",
    LpStatus[problem.status]
)

# Optimal values
print(
    "Product A Units:",
    A.varValue
)

print(
    "Product B Units:",
    B.varValue
)

print(
    "Maximum Profit = ₹",
    value(problem.objective)
)

print("""
Business Insights

1. Produce 20 units of Product A.
2. Produce 60 units of Product B.
3. Maximum achievable profit is ₹2600.
4. All available resources are efficiently utilized.
5. Optimization helps improve business profitability.
""")
```
